```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os 
import sys
```

```python
d1 = pd.read_csv("../data/raw/PTF_SMF_ve_SDF_Listeleme-01012024-01012025.csv",sep=";")  #data from 2024
d2 = pd.read_csv("../data/raw/PTF_SMF_ve_SDF_Listeleme-01012025-20122025.csv",sep=";")  #data from 2025
epias_data = pd.concat([d1,d2], ignore_index=True)                                #combine data 2024-2025

epias_data.head()
```

```python
epias_data.info()
```

```python
null_data = epias_data[epias_data.isnull().any(axis=1)]

display(null_data)
```

```python
epias_data.dropna(inplace=True)
```

```python
epias_data.describe().T
```

```python
num_cols = epias_data.columns.drop(['Tarih', 'Saat', "SMF Yön"])
for col in num_cols:
    epias_data[col] = epias_data[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
    epias_data[col] = pd.to_numeric(epias_data[col])
```

```python
epias_data.describe().T
```

```python
epias_data.head(20)
```

```python
epias_data.reset_index(drop=True, inplace=True)
epias_data.index = pd.to_datetime(epias_data['Tarih'] + ' ' + epias_data['Saat'], dayfirst=True)

epias_data.head()
```

```python
epias_data.index
```

```python
epias_data["month"] = epias_data.index.month
epias_data["day"] = epias_data.index.day_name()


epias_data.loc[epias_data['month'].isin([12, 1, 2]), 'season'] = 'winter'
epias_data.loc[epias_data['month'].isin([3, 4, 5]), 'season'] = 'spring'
epias_data.loc[epias_data['month'].isin([6, 7, 8]), 'season'] = 'summer'
epias_data.loc[epias_data['month'].isin([9, 10, 11]), 'season'] = 'autumn'
```
